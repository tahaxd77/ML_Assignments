```python
# 28/05/2025
# CSC354 â€“ Assignment1 â€“ ML â€“ Concept Learning
# Nimra Shahid
# FA22-BCS-123
# Applying Candidate Elimination Method on the given data set



import numpy as np
import pandas as pd
from sklearn.model_selection import KFold

def get_most_general_hypothesis(num_attributes):
    return ["?"] * num_attributes

def get_most_specific_hypothesis(num_attributes):
    return ["âˆ…"] * num_attributes

def consistent(h, instance):
    for i in range(len(h)):
        if h[i] != "?" and h[i] != instance[i]:
            return False
    return True

def candidate_elimination(training_data):

    num_attributes = len(training_data[0]) - 1
    S = None
    G = [get_most_general_hypothesis(num_attributes)]

    for instance in training_data:
        x = instance[:-1]
        y = instance[-1]

        if y == "English":
            if S is None:
                S = x[:]
            else:
                S = [x[i] if S[i] == x[i] else "?" for i in range(num_attributes)]
            G = [g for g in G if consistent(g, x)]

        else:
            G = [g for g in G if not consistent(g, x)]
            new_G = []
            for g in G:
                for i in range(num_attributes):
                    if g[i] == "?":
                        new_h = g[:]
                        new_h[i] = S[i]
                        new_G.append(new_h)
            G += new_G

    return S, G

file_path = "books_data-Final.xlsx"
df = pd.read_excel(file_path)


data = df.iloc[:, 1:].values.tolist()

kf = KFold(n_splits=10, shuffle=True, random_state=42)
fold_results = []

for fold, (train_idx, test_idx) in enumerate(kf.split(data), 1):
    train_data = [data[i] for i in train_idx]
    test_data = [data[i] for i in test_idx]

    S_final, G_final = candidate_elimination(train_data)

    print(f"\nðŸ”¹ Fold {fold}")
    print("Final Specific Hypothesis:", S_final)
    print("Final General Hypothesis Set:", G_final)

    fold_results.append((S_final, G_final))

```
