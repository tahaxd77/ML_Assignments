```python


import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Distance matrix
labels = ['P1', 'P2', 'P3', 'P4', 'P5']
D = np.array([
    [0.00, 0.10, 0.41, 0.55, 0.35],
    [0.10, 0.00, 0.64, 0.47, 0.98],
    [0.41, 0.64, 0.00, 0.44, 0.85],
    [0.55, 0.47, 0.44, 0.00, 0.76],
    [0.35, 0.98, 0.85, 0.76, 0.00]
])

# Convert to condensed form
condensed_D = sch.distance.squareform(D)

# SINGLE LINKAGE
Z_single = sch.linkage(condensed_D, method='single')
plt.figure(figsize=(8, 4))
sch.dendrogram(Z_single, labels=labels)
plt.title("Single Linkage Dendrogram")
plt.show()

# COMPLETE LINKAGE
Z_complete = sch.linkage(condensed_D, method='complete')
plt.figure(figsize=(8, 4))
sch.dendrogram(Z_complete, labels=labels)
plt.title("Complete Linkage Dendrogram")
plt.show()
```
